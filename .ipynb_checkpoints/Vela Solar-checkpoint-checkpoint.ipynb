```python
def oi_amanda(ci, t):
    x = ci[0]
    vx = ci[1]
      
    forca = potencia / c

    Rx = forca * area
    
    dxdt = vx * P/L
    dvxdt = Rx / m  * (P**2/L)
    
    global aceleracao
    aceleracao.append(dvxdt)

    return dxdt, dvxdt
```

```python
from scipy.integrate import odeint
import numpy as np
import matplotlib.pyplot as plt

aceleracao = []

L = 1.5 * 10*11
P = 3.15 * 10**7

x0 = 1
vx0 = 50 * (L/P)
potencia = 50 * 10**6
ci = [x0,vx0]
c = 3 * 10**8 * (L/P)
lado = 200 
area = lado**2 * L**2
m = 100 #kg

listaTempo = np.arange(0,30,0.001)
    
resp = odeint(oi_amanda, ci, listaTempo)

plt.figure(figsize=(7,7))
plt.plot(listaTempo, resp[:,0])
plt.xlabel('Tempo(anos)')
plt.ylabel('Posicao')
plt.title('posicao')
plt.grid(True)
plt.show()

plt.figure(figsize=(7,7))
plt.plot(listaTempo, resp[:,1])
plt.xlabel('Tempo(anos)')
plt.ylabel('Velocidade')
plt.title('vel')
plt.grid(True)
plt.show()

plt.figure(figsize=(7,7))
plt.plot(range(len(aceleracao)), aceleracao, 'bo')
plt.xlabel('Tempo (anos)')
plt.ylabel('Aceleração')
plt.title('a')
plt.grid(True)
plt.show()
```
